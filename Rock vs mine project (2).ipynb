```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import seaborn as sns
```

```python
sonar_data=pd.read_csv('sonar.csv',header=None)
sonar_data.head()
```

```python
sonar_data.shape
```

```python
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sns.countplot(data=sonar_data,x=sonar_data[60])
```

```python
sonar_data.groupby(60).mean()
```

```python
x=sonar_data.drop(columns=60,axis=1)
y=sonar_data[60]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
```

```python
accuracy_x_train =accuracy_score(x_train_prediction,y_train)
accuracy_x_train
```

```python
x_test_prediction=model.predict(x_test)
accuracy_x_test=accuracy_score(x_test_prediction,y_test)
accuracy_x_test
```
